##### Import packages

```python
import cvxpy as cp
import numpy as np
import matplotlib.pyplot as plt

from ampyc.utils import Polytope
```

##### Usage of the Polytope class

```python
# define a polytope in halfspace representation
A_x = np.array([[1, 0], [-1, 0], [0, 1], [0, -1], [1, 1]])
b_x = np.array([[1], [1], [2], [2], [2]])
X = Polytope(A_x, b_x)
```

```python
# plot the polytope
_, ax = plt.subplots(1,1, figsize=(5,4))
X.plot(ax, color="gray", alpha=1, edgecolor='black', linewidth=2)

# you can use the provided xlim and ylim properties to set the plot limits
ax.set_xlim(X.xlim)
ax.set_ylim(X.ylim)

plt.show()
```

```python
# get the vertices of the polytope
print("vertices\n{0}\n".format(X.V))

# get halfspace representation
print("halfspace representation\n{0}\t{1}\n".format(X.A, X.b))

# get dimension of the polytope
print("dimension: {0}\n".format(X.dim))

# get volume of the polytope
print("volume: {0}\n".format(X.volume))
```

```python
# scale the polytope
X_scaled = 0.5 * X

# apply a linear transformation to the polytope
A = np.array([[1, 0.2], [0, 1]])
X_transformed = A @ X

# plot the polytopes
_, ax = plt.subplots(1,1, figsize=(5,4))
X.plot(ax, color="gray", alpha=0.8, edgecolor='black', linewidth=2, label="original")
X_scaled.plot(ax, color="blue", alpha=0.3, edgecolor='blue', label="scaled")
X_transformed.plot(ax, color="green", alpha=0.3, edgecolor='green', label="transformed")

# you can use the provided xlim and ylim properties to set the plot limits
ax.set_xlim(X_transformed.xlim)
ax.set_ylim(X_transformed.ylim)

plt.legend(loc="upper left")
plt.show()
```

```python
# define another polytope
W = Polytope(A=np.array([[1, 0], [0, 1], [-1, 0], [0, -1]]), b=np.array([[0.2], [0.2], [0.2], [0.2]]))

# Minkowski sum
X_minkowski = X + W

# Pontryagin difference
X_pontryagin = X - W

# plot the polytopes
_, ax = plt.subplots(1,1, figsize=(5,4))
X.plot(ax, color="gray", alpha=0.8, edgecolor='black', linewidth=2, label="original")
X_minkowski.plot(ax, color="blue", alpha=0.25, edgecolor='darkblue', label="Minkowski")
X_pontryagin.plot(ax, color="green", alpha=0.25, edgecolor='darkgreen', label="Pontyagin")
(X.V[-1] + W).plot(ax, color="red", alpha=0.25, edgecolor='darkred', label="W")

# you can use the provided xlim and ylim properties to set the plot limits
ax.set_xlim(X_minkowski.xlim)
ax.set_ylim(X_minkowski.ylim)

plt.legend(loc="upper right")
plt.show()
```

##### Quadratic Programming with cvxpy

```python
# define dynamic matrices
A = np.array([[1, 0.5], [0.2, 1]])
B = np.array([[0], [0.1]])

# define state and input constraints
A_x = np.array([[1, 0], [-1, 0], [0, 1], [0, -1], [1, 1]])
b_x = np.array([[1], [1], [2], [2], [2]])
X = Polytope(A_x, b_x)

# define cost matrices
Q = np.diag([1,0.5])
R = 10 * np.eye(1)

# define optimization variables
x = cp.Variable((2, 6))
u = cp.Variable((1, 5))
x_0 = cp.Parameter((2,))

# define objective
obj = 0
for i in range(5):
    obj += cp.quad_form(x[:,i], Q) + cp.quad_form(u[:,i], R)

# define constraints
constraints = [x[:,0] == x_0]
for i in range(5):
    constraints += [x[:,i+1] == A @ x[:,i] + B @ u[:,i]]
    constraints += [X.A @ x[:,i] <= X.b]
constraints += [x[:,5] == 0.0]

# define optimization problem
prob = cp.Problem(cp.Minimize(obj), constraints)

# simulate for 30 time steps
x_k = np.zeros((2, 30))
x_k[:,0] = np.array([0, 1.8])

for i in range(29):
    x_0.value = x_k[:,i]
    prob.solve()
    x_k[:,i+1] = A@x_k[:,i] + B@u[:,0].value
```

```python
# plot the simulated trajectory
_, ax = plt.subplots(1,1, figsize=(5,4))
X.plot(ax, fill=False, alpha=0.8, edgecolor='black', linewidth=2, label="X")

ax.plot(x_k[0], x_k[1], color="blue", label="trajectory")

# you can use the provided xlim and ylim properties to set the plot limits
ax.set_xlim(X.xlim)
ax.set_ylim(X.ylim)

plt.legend(loc="lower left")
plt.grid(True)
plt.show()
```

##### Semi-Definite Programming with cvxpy

```python
# define state and input constraints
A_x = np.array([[1, 0], [-1, 0], [0, 1], [0, -1], [1, 1]])
b_x = np.array([[1], [1], [2], [2], [2]])
X = Polytope(A_x, b_x)

# define optimization variable
E = cp.Variable((2, 2), symmetric=True)

# define objective
obj = -cp.log_det(E)

# define constraints
constraints = [E >> 0] # The operator >> denotes a definitness constraint
for i,A_i in enumerate(X.A):
    constraints += [cp.quad_form(A_i, E) <= X.b[i]**2]

# define optimization problem
prob = cp.Problem(cp.Minimize(obj), constraints)

# solve optimization problem
prob.solve()
```

```python
# extract optimal solution
P = np.linalg.inv(E.value)

# print result
print("optimal P:\n{0}\n".format(P))

# plot result
_, ax = plt.subplots(1,1, figsize=(5,4))
X.plot(ax, fill=False, alpha=0.8, edgecolor='black', linewidth=2, label="X")

L = np.linalg.cholesky(P)
t = np.linspace(0,2*np.pi,200)
z = np.vstack([np.cos(t), np.sin(t)])
ell = np.linalg.inv(L.T)@z

ax.plot(ell[0,:], ell[1,:], color="blue", label="P")

# you can use the provided xlim and ylim properties to set the plot limits
ax.set_xlim(X.xlim)
ax.set_ylim(X.ylim)

plt.legend(loc="lower left")
plt.grid(True)
plt.show()
```
